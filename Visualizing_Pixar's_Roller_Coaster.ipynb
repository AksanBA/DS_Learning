```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #Pretty Data Viz
%matplotlib inline 
#Plots are displayed inline

pixar_movies=pd.read_csv('PixarMovies.csv') #Reading in file

rows=pixar_movies.shape[0] #returns number of rows
columns=pixar_movies.shape[1] #returns number of columns

print(pixar_movies.head(rows)) #will display the entire dataframe

print(pixar_movies.dtypes) #displays types
pixar_movies.describe() #summary stats
```

```python
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies['International %']=pixar_movies['International %'].str.rstrip('%').astype('float')


pixar_movies.head(rows)
```

```python
pixar_movies['IMDB Score']=pixar_movies['IMDB Score']*10
#makes all critic scores out of 100 scale

filtered_pixar=pixar_movies.loc[0:13] #takes the first 14 rows, ignores Inside Out

pixar_movies.set_index('Movie',inplace=True)
filtered_pixar.set_index('Movie',inplace=True)

filtered_pixar.head(14)
```

```python
critics_reviews=pixar_movies[['RT Score', 'IMDB Score','Metacritic Score']]

critics_reviews.plot(figsize=(10,6))
plt.show()
```

```python
critics_reviews.plot(kind='box',figsize=(9,5))
plt.show()
```

```python
revenue_proportions=filtered_pixar[['Domestic %','International %']]
revenue_proportions.plot(kind='bar',stacked=True)
plt.show()
```
