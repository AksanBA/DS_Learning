```python
import pandas as pd

police_killings = pd.read_csv('police_killings.csv', encoding = 'ISO-8859-1')

print(police_killings.columns)
police_killings.head()
```

```python
races = police_killings['raceethnicity'].value_counts()
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
fig = plt.figure()
plt.xticks(np.arange(len(races)),races.index)
plt.bar(np.arange(len(races)), races)
```

```python
race_percentage = pd.Series({'White': 63.7, 'Black': 12.2, 'Hispanic/Latino': 16.3, 'Asian/Pacific Islander': 4.9, 'Native American': 0.7, 'Unknown': 2.1}
                    )
race_total = (race_percentage/100) * police_killings.shape[0]
race_total
```

```python
killing_ratio = races / race_total
fig = plt.figure()
plt.xticks(np.arange(len(races)),races.index)
killing_ratio.plot(kind='bar')
```

Black people are disproportionately killed compared to other races using census information of their population percentage.

```python
income = police_killings[police_killings['p_income'] != '-']['p_income'].astype(int)
income.hist(bins=20)
```

The median US income is about $50000 so this histogram indicates that killings are happening in lower income areas.

```python
state_pop = pd.read_csv('state_population.csv')
counts = police_killings['state_fp'].value_counts()

states = pd.DataFrame({"STATE": counts.index, "shootings": counts})
```

```python
states = states.merge(state_pop, on = 'STATE')

states
```

```python
states['pop_millions'] = states.POPESTIMATE2015 / 1000000
states['rate'] = states.shootings / states.pop_millions

statessort = states.sort_values('rate', ascending = 0)
```

```python
pk = police_killings[(police_killings['share_white'] != '-') &
                     (police_killings['share_black'] != '-') &
                     (police_killings['share_hispanic'] != '-')]
pk[['share_white','share_black','share_hispanic']] = pk[['share_white','share_black','share_hispanic']].astype('float')

pk[['share_white','share_black','share_hispanic']].dtypes
```

```python
top10 = ['OK', 'AZ', 'NE', 'HI', 'AK', 'ID', 'NM', 'LA', 'CO', 'DE']
bottom10 = ['CT', 'PA', 'IA', 'NY', 'MA', 'NH', 'ME', 'IL', 'OH', 'WI']
top10n = statessort[:10]['STATE']
bottom10n = statessort[-10:]['STATE']
#pk['state_fp']
```

```python
pktop10 = pk[pk['state'].isin(top10)]
pkbottom10 = pk[pk['state'].isin(bottom10)]
```

```python
pktop10[['pop', 'county_income', 'share_white', 'share_black', 'share_hispanic']].mean()
```

```python
pkbottom10[['pop', 'county_income', 'share_white', 'share_black', 'share_hispanic']].mean()
```

The results show a conistent population area for both the bottom and top state shooting rates. There is a noticeable difference in financial inequality between the top shooting rate states and lowest ones.

Interestingly, the states that have the highest shooting rates actually have a higher hispanic population share rather than a higher black population share.

There may be a couple of factors at play here. Removing the rows with a '-' may have skewed results. Also, the averages the last two cells are for county level data rather than state level.

Because we're looking at county level this could also skew the race data. For example a shooting of a minority might occur in a neighbouring county where the minority population is different but they live in another county.
